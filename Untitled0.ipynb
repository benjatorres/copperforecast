Descripción de variables
-	copper: precio spot del cobre, dólares por tonelada métrica
-	gold: precio spot del oro, dólares por onza
-	silver: precio spot de la plata, dólares por onza
-	iron: precio spot del hierro, dólares por tonelada métrica
-	oil: precio del petróleo West Texas Intermediate (WTI) en dólares por barril
-	copper_L1: primer rezago precio del cobre
-	copper_L2: dos rezagos del precio del cobre
-	copper_L3: tres rezagos del precio del cobre
-	copper_L4: cuatro rezagos del precio del cobre
-	copper_L5: cinco rezagos del precio del cobre
-	activity_L1: rezago del índice de actividad económica global
-	aus: tipo de cambio (dólares australianos por dólar estadounidense)
-	chn: tipo de cambio (yuanes chinos por dólar estadounidense)
-	clp: tipo de cambio (pesos chilenos por dólar estadounidense)
-	per: tipo de cambio (soles peruanos por dólar estadounidense)
-	date: periodo enero 1990- marzo 2023
- id: identificador de observaciones

```python
#Rezagamos las variables
data['gold_lag_1'] = data['gold'].shift(1)
data['gold_lag_2'] = data['gold'].shift(2)
data['gold_lag_3'] = data['gold'].shift(3)
data['gold_lag_4'] = data['gold'].shift(4)
data['gold_lag_5'] = data['gold'].shift(5)

data['iron_lag_1'] = data['iron'].shift(1)
data['iron_lag_2'] = data['iron'].shift(2)
data['iron_lag_3'] = data['iron'].shift(3)
data['iron_lag_4'] = data['iron'].shift(4)
data['iron_lag_5'] = data['iron'].shift(5)

data['oil_lag_1'] = data['oil'].shift(1)
data['oil_lag_2'] = data['oil'].shift(2)
data['oil_lag_3'] = data['oil'].shift(3)
data['oil_lag_4'] = data['oil'].shift(4)
data['oil_lag_5'] = data['oil'].shift(5)

data['silver_lag_1'] = data['silver'].shift(1)
data['silver_lag_2'] = data['silver'].shift(2)
data['silver_lag_3'] = data['silver'].shift(3)
data['silver_lag_4'] = data['silver'].shift(4)
data['silver_lag_5'] = data['silver'].shift(5)

data['aus_lag_1'] = data['aus'].shift(1)
data['aus_lag_2'] = data['aus'].shift(2)
data['aus_lag_3'] = data['aus'].shift(3)
data['aus_lag_4'] = data['aus'].shift(4)
data['aus_lag_5'] = data['aus'].shift(5)

data['chn_lag_1'] = data['chn'].shift(1)
data['chn_lag_2'] = data['chn'].shift(2)
data['chn_lag_3'] = data['chn'].shift(3)
data['chn_lag_4'] = data['chn'].shift(4)
data['chn_lag_5'] = data['chn'].shift(5)

data['per_lag_1'] = data['per'].shift(1)
data['per_lag_2'] = data['per'].shift(2)
data['per_lag_3'] = data['per'].shift(3)
data['per_lag_4'] = data['per'].shift(4)
data['per_lag_5'] = data['per'].shift(5)
```

```python
data['date'] = pd.to_datetime(data["date"].dt.strftime('%Y-%m'))
data.head()
```

```python
import yfinance as yf

# Read your existing dataset with dates
data['date'] = str(pd.to_datetime(data["date"].dt.strftime('%Y-%m')))

# Extract the unique dates from your dataset
dates = data['date'].unique()

# Create an empty list to store the S&P 500 index values
sp500_values = []

# Fetch the S&P 500 index values for each date
for date in dates:
    sp500_data = yf.download('^GSPC', start=date, end=date)

    # Check if data is available for the specified date
    if not sp500_data.empty:
        sp500_value = sp500_data['Close'].values[0]
        sp500_values.append(sp500_value)
    else:
        sp500_values.append(None)  # Add None for missing data
# Add the S&P 500 index values to your dataset
data['sp500_value'] = sp500_values
```

**RANDOM FOREST**

```python
# Importamos librerías
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# La variable dependiente del modelo es copper, y descartamos las demás variables que no necesitaremos.
# Además, eliminamos filas con missings dado que RF no permite utilizarlas.
data_sin_missings = data.drop(range(5))
rf_X = data_sin_missings.drop(['copper', 'date', 'activity', 'id'], axis=1)
rf_y = data_sin_missings['copper']
```

```python
# Dividimos el dataset en entrenamiento y testeo
rf_X_train, rf_X_test, rf_y_train, rf_y_test = train_test_split(rf_X, rf_y, test_size=0.2, random_state=12)
rf_X_train.shape, rf_X_test.shape
```

```python
# Creamos el "bosque" usando criterio mse
rf = RandomForestRegressor(random_state=12, criterion='squared_error')
```

```python
# Entrenamos el modelo

rf.fit(rf_X_train, rf_y_train)
```

```python

plt.plot(rf_y_test, marker='o', label='Valores reales')
```
